```python
def print_prime() :
    l = []
    for i in range(1,1000) :
        c = 0
        for j in range(2,i) :
            if i % j == 0 :
                c = 1
        if c != 1 :
            l.append(i)
    return l
        
print_prime()    
```

```python
def myprint(*S,end = '\n',sep = ' ') :
    p = ''''''
    for i in S:   
        p += str(i)
        p += sep         # separator for every separated value by , while passing multiple values through S.
    p += end          
    return p


myprint(["ankur"],"yadav",{'a':23, 2:[1,2,3]})
```

```python
def myappend(l,S):
    l += [S]
    return l

def myextend(l,S):
    """l is list which need to be extended with values S"""
    if type(S) in (str,list,tuple,set,dict) and type(l) == list:
        for i in S:
            l += [i] # performing simple addition of 2 lists by typecasting iterable objects
        return l
    else :
        return "Please pass valid parameters"
        

def pop01(l, index = -1):
    """ pop01(l = list, index(Bydefault = -1))"""
    if type(l) == list and index != -1 :
        l = l[0:index] + l[index+1:]   # it will not consider value at provided index
    else :
        l = l[:-1]
    return l
```

```python
# filter even values
l1 = lambda *x : [i for i in x if i%2 ==0 ]
l1(1,2,3,4,5,6,7,8)
```

```python
# to get squares and cubes in a dictionary
l2 = lambda *x : { 'square':[i**2 for i in x ] , 'cube':[i**3 for i in x]}
l2(2,3,4,5)
```

```python
# adding each element of 2 lists, does not returns elements which do not have counter element to add.
l3 = lambda x,y : [i + j for i,j in zip(x,y) if type(x) == type(y) == list]
l3([2,4,7],[6,5,4,7,8])
```

```python
# repeats values specific no. of times for string, 2nd parameter should be integer
l4 = lambda x,y : [i*j for i in x for j in range(1,y+1) ]
l4(['a','b'],3)
```

```python
# removes duplicate values and returns ordered list
l5 = lambda x : list(set(x))
l5([4,3,2,5,3,4,3,"as",2.345])
```

```python
# returns index of vowels in a string.
l6 = lambda x : [ i for i in range(len(x)) if x[i].lower() in "aeiou"]
l6("I am writing lambda function")
```

```python
# returns index where a string contains sub-tring 
l7 = lambda x,y : [i for i in range(1,len(x)-len(y)) if x[i:i+len(y)] == y ]
l7("your name is __ is this correct?","is")
```

```python
# convert list containing lists,tuple,set and string to flat list
l8 = lambda x : [j for i in x for j in i ]
l8([[1,2,3],('a','v'),{8,9}])
```

```python
# converts dictionary into list with key:value pair as a single element
l9 = lambda x : [str(i) + " : " + str(j) for i,j in list(x.items())]
l9({1:2,'a':'b'})
```

```python
# removes spaces in a string
l10 = lambda x : ''.join([i for i in x if i != ' '])
l10("hello how are you")
```
